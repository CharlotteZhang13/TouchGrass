```python
import requests
from bs4 import BeautifulSoup

# Function to get HTML content from a URL
def get_html(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Check if the request was successful
    return response.text

# Function to parse HTML and extract elements
def scrape_elements(html):
    soup = BeautifulSoup(html, 'html.parser')
    image_sources = []
    # Find all <a> elements with the specified aria-label
    links = soup.find_all('a', {'aria-label': 'Image result for dormitory'})
    for link in links:
      img_tag = link.find('img')
      if img_tag and 'data-src' in img_tag.attrs:
        image_sources.append(img_tag['data-src'])
    return image_sources


# URL to scrape
url = "https://www.bing.com/images/search?q=dormitory&form=HDRSC3&first=1"
# Get HTML content
html = get_html(url)

# Scrape and print the elements
elements = scrape_elements(html)
for element in elements:
    print(element)

import os
import requests

def download_image(url, folder_path):
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()  # Check if the request was successful
        file_name = url.split("/")[-1]
        file_path = os.path.join(folder_path, file_name)

        # Save the image to the specified path
        with open(file_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)

        print(f"Downloaded {file_name}")
    except Exception as e:
        print(f"Error downloading {url}: {e}")

# Create a folder in Google Drive to store images
folder_path = '/content/drive/MyDrive/Images'
os.makedirs(folder_path, exist_ok=True)
for url in elements:
  download_image(url, folder_path)
```

```python
from PIL import Image
import os

def convert_to_jpg(input_path, output_path):
    try:
        # Open the image file
        with Image.open(input_path) as img:
            # Convert image to RGB (necessary for JPEG format)
            rgb_img = img.convert('RGB')

            # Save the image as a .jpg file
            rgb_img.save(output_path, 'JPEG')

        print(f"Image saved as {output_path}")
    except Exception as e:
        print(f"Error converting image {input_path}: {e}")

# Directory containing the images
input_directory = '/content/drive/MyDrive/Images'
output_directory = '/content/drive/MyDrive/Images/JPG'

# Create the output directory if it does not exist
os.makedirs(output_directory, exist_ok=True)

# Iterate through all files in the directory
for file_name in os.listdir(input_directory):
    file_path = os.path.join(input_directory, file_name)

    # Ensure it's a file (not a directory) and has a valid extension
    if os.path.isfile(file_path) and not file_name.lower().endswith('.jpg'):
        # Define the output file path
        output_file_name = os.path.splitext(file_name)[0] + '.jpg'
        output_file_path = os.path.join(output_directory, output_file_name)

        # Convert the image to JPG
        convert_to_jpg(file_path, output_file_path)
```
